# Dropna and Fillna

```python
import pandas as pd
```

```python
var = pd.read_csv("sample1.csv")
var
```

```python
var1 = pd.read_csv("sample1.csv",nrows = 50)
var1
```

```python
var1.dropna() #---------------the dropna worked along the row ..................
```

#### Dropna Along the column ------

```python
var1.dropna(axis =1)   # here .... axis 1 means the column
```

### how = any/all function in dropna 

```python
var1.dropna(how = "any", axis =1)  
```

```python
var1.dropna(how = "all", axis =1)  
```

# Note : 
##### how = "any" means , it drops all values if a row/column contains one or more  NaN
##### how = "all" means , it will drop the row / column if all values of a specific row / column is NaN

```python
# Subset parameter 

var1.dropna(subset = ["Period"]) # removes NaN value of a particular column 
```

```python
# inplace parameter 

var1.dropna(inplace = True)  # creates a new data set removing all NaN values


```

```python
# thresh parameter 

var.dropna(thresh =1 ) # removes n = 1 number of NaN values 

```

# Fillna starts from here 


```python
# basic of fillna 

var.fillna("sudeep") # replace NaN with string "sudeep"

```

```python

# replaces NaN values specific to the particular column  as shown in below using dictionary 


var.fillna({"Series_reference": "shreya", "Period": "love", "Data_value":"sudeep"})
```

```python


# --------forward filing and backward filling -----------


# --var.fillna(method = "ffill") # for forward filling 

var.fillna(method = "bfill" ,axis = 1) # for backward filling 
 # also axis can be passed which mean data can be filled horizontally too 
    
```

```python
# inplce parameter 


var.fillna(12, inplace= True)
var
```

```python
# limit parameter __ n number of NaN is replaced !

var.fillna("sudeep",limit = 1 )


    
```
