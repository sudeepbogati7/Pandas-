# melt()


##### - > used to reshape the data items or tables 
##### -> represent data at a single column 


```python
import pandas as pd
```

```python
score = pd.DataFrame({"Day": [1,2,3,4,5,6,7], "Eng": [21,22,45,65,78,98,45], "Maths":[56,78,98,12,10,45,65], "Nep":[55,65,85,75,65,95,45]})
score
```

```python
# using melt function 
pd.melt(score)
```

```python
pd.melt(score,var_name  = "Subjects", value_name = "Marks" )
```

#### That's it about melt() ! 


# Pivot()

```python
var = pd.DataFrame({"Day": [1,2,3,4,5,6,7],"St_name": ["Shrinika", "Sudeep","Bibek","Anurag","Saugat","Binod","Sanjiv"], "Eng": [21,22,45,65,78,98,45], "Maths":[56,78,98,12,10,45,65], "Nep":[55,65,85,75,65,95,45]})
var
```

```python
var.pivot(index = "Day", columns = "Eng")
```

```python
var.pivot_table(index = "St_name", columns = "Eng", 
               aggfunc = "mean")
```

```python
var.pivot_table(index = "St_name", columns = "Eng", 
               aggfunc = "sum")
```
