```python
import pandas as pd
```

# Merge

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B": [6,7,8,9,10]})
var2=pd.DataFrame({"A":[1,2,3,4,6], "C": [6,56,5,6,0]})
```

```python
pd.merge(var1,var2, on= "A")
```

```python
# changing the order while passing parameter, changes the order of data 


pd.merge(var2,var1, on= "A")
```

```python
# how parameter

pd.merge(var2,var1, how = "outer")
pd.merge(var2,var1, how = "left")
pd.merge(var2,var1, how = "right")
```

```python
pd.merge(var2,var1, how = "outer", indicator = True)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B": [6,7,8,9,10]})
var2=pd.DataFrame({"A":[1,2,3,4,6], "B": [6,56,5,6,0]})

pd.merge(var1,var2, left_index= True, right_index = True)
```

```python
# ------- suffix ---------

pd.merge(var1,var2, left_index= True, right_index = True,suffixes =("-  name"," -roll no"))
```

# Concat

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([5,6,7,8])
```

```python
# -------- series concat------------


pd.concat([sr1,sr2])
```

```python
# ------- DataFrame concat -----------

d1 = pd.DataFrame({"A":[1,2,3,4,5], "B": [6,7,8,9,10]})
d2 = pd.DataFrame({"C":[1,2,5], "D": [6,6,0]})

cc = pd.concat([d1,d2])
cc

```

```python
cc = pd.concat([d1,d2], axis = 1)
cc
```

```python
cc = pd.concat([d1,d2], keys = ["d1","d2"])
cc
```
