### Sort in ascending order 

```python
csv_1.sort_index(axis = 0, ascending = False)

# Note : axis =0 means the row and axis = 1 means according to column
```

### change the specific data (LOC FUNCTION)

```python
csv_1.loc[0,"Footnote"] = "Sudeep "
csv_1
```

```python
csv_1.loc[[0,1],["Footnote"]]    

# this displays the specific data of certain row and column
```

```python
csv_1.loc[[0,1],:]
```

```python
csv_1.loc[:,["Footnote"]]
```
