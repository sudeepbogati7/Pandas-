```python
var.interpolate() # showing error 'cause it works only with number ----
```

# Note: 
### interpolate() function works only with the numerical value or numbers 

## Methods are used to change the data filling style 


# Methods : 

#### {linear, time, index, values, nearest, zero, slinear, quadratic, cubic, barycentric, krogh , polynimial, spline, piecewise_polynomial, form_derivatives, pchip, akima }

```python
var.interpolate(method = "linear", axis = 1) 
```

## axis = 1 :  row by row 

## axis = 0 :   column by column

```python
# ------- limit parameter =-------

# limit _direction = "forward"/"backward"/ "both"


var.interpolate(limit_direction = "forward",limit = 1)
var.interpolate(limit_direction = "backward",limit = 1)
var.interpolate(limit_direction = "both",limit = 1)
```

```python
 # ------------limit area ------
var.interpolate(limit_area = "outside")   # do not fill the NaN values 
var.interpolate(limit_area = "inside")
```

```python
#----------inplace parameter --------
# inplace = True : original data placement
# inplace = False : copies the data and fill

var.interpolate(inplace = True, limit_direction = "both ", limit = 1)
```
