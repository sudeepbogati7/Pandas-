 # Read CSV

```python
import pandas as pd
```

```python
csv_1 = pd.read_csv("test.csv")  #reads the csv file
csv_1
```

```python
csv_2 = pd.read_csv("test.csv",nrows= 100) #to get n number of rows
csv_2
```

```python
csv_4 = pd.read_csv("test.csv",nrows= 10) #to get n number of rows
csv_4
```

```python
csv_4 = pd.read_csv("test.csv", usecols= ["Industry_code_ANZSIC06"]) #to get specific column
csv_4
```

```python
csv_5 = pd.read_csv("test.csv", usecols= ["Industry_code_ANZSIC06","Year","Units"]) #to get specific columns
csv_5
```

### get specific columns by passing index number

```python
csv_6 = pd.read_csv("test.csv", usecols= [0,3,5,4,5]) #using index number 
csv_6
```


## Skip a specific row

```python
csv_4 = pd.read_csv("test.csv",nrows= 10, skiprows = [0,1]) 
csv_4
```

# make index of a column

```python
csv_7 = pd.read_csv("test.csv", nrows=5, index_col = "Year") #changes the index to "Year" column
csv_7
```

### Change the header

```python
csv_8 = pd.read_csv("test.csv", nrows=5, header = 5) #header is changed .... the fifth row is header now !
csv_8
```

# Add or changes custom heading (name function)

```python
csv_9 = pd.read_csv("test.csv", nrows=5, names = ["a","b","c","d","e","f","g","h","i","j"]) 
csv_9
```

## Use of prefix in the heading 

```python
csv_10 = pd.read_csv("test.csv", nrows=5, , header = None, prefix = "col")
csv_10
```
