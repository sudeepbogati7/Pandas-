```python
import pandas as pd 
```

```python
var = pd.read_csv("test.csv")
var


#------------replace --------------------
```

```python
var.replace(to_replace = "Level 1", value = 22)
```

```python
# replace using array 

var.replace([2021, "Level 1"], 0)
```

```python
# replace the alphabetical words 

var.replace("[A-Za-z]","sudeep", regex= True)
```

```python
# replace using dictionary

var.replace({"Industry_code_ANZSIC06": '[a-z]' },12, regex = True )
```

```python
# -----------method parameter -------------

#var.replace(1, method = "ffill")
var.replace(1, method = "bfill")
```

```python
# -------passing limit as a parameter ------------


var.replace(7, method= "ffill", limit = 1) #only replaces one data 
```

```python
var.replace(7, method= "ffill", limit = 1, inplace = True)#replaces the original data set without making the copy 
var
```
